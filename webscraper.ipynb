Imports

```python
import requests
from bs4 import BeautifulSoup
import urllib
# for table analysis
import pandas as pd
# write to csv
import csv
import random
```

Get list of 50 most common russian nouns

```python
URL = "http://masterrussian.com/vocabulary/common_nouns.htm"
page = requests.get(URL)

words = BeautifulSoup(page.content, "html.parser")
word_list = [d.text.rstrip() for d in words.find_all("td",{"class":"word"})]
print(word_list)
```

For word in list, get declension information on word

```python
flag = True

for i in range(len(word_list)):
  URL = "https://en.wiktionary.org/wiki/{f}#Russian".format(f=word_list[i])
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, "html.parser")
  head = soup.find('span',{"lang":"ru","class":"form-of"})
  right_table = head.find_parent("tbody")

  for row in right_table.findAll("tr"):
    cells = row.findAll('td')
  rows = right_table.findAll("tr",limit=7)

  #make column indices
  if flag:
    header = [th.text.rstrip() for th in rows[0].find_all('th')][1:]
    cases = [row.find('th').text.rstrip() for row in rows[1:]]
    multicol = pd.MultiIndex.from_product([cases,header])
    df = pd.DataFrame(columns=multicol)
    df.insert(0,'declension',[])
    flag = False
  else:
    pass

  #get declension data
  declension = soup.find('a',{'href':"/wiki/Appendix:Russian_nouns#Declension_tables"}).text.rstrip()
  declension = declension.strip('[]').split()
  lst_data1 = [declension]
  for row in rows[1:]:
    data = [d.text.rstrip() for d in row.find_all("span",{"lang":"ru"},limit=2)]
    lst_data1 += data
  try:
    df.loc[i] = lst_data1
  except:
    pass





```
